```python
ru_emb = KeyedVectors.load_word2vec_format("cc.ru.300.vec")
```

```python
tokenizer = nltk.tokenize.TweetTokenizer()
morph = pymorphy2.MorphAnalyzer()
```

```python
def word_prepare(sentence):
    bad_words = {'кульверт': 'труба', 'панафриканским': 'африканский', 'столицой': 'столицей', 'тюрскской': 'турецкой',
                 'АДС-М': 'прибор', 'оттоманка': 'диван', 'менделеева': 'Менделеева', 'овцевода': 'пастуха', 
                 'Хэмингуэя': 'Хэмингуэй', 'крестражем': 'фантастика', 'школьница-подросток': 'школьница', 
                 'Хворостухина': 'цифры', 'телескор': 'телескоп', 'клеверсы': 'валюта', 'видущих': 'ведущих',
                }
    words_dict = tokenizer.tokenize(sentence)
    l = []
    for word in words_dict:
        if word in bad_words.keys():
            l.append(ru_emb[bad_words[word]])
        else:
            try:
                l.append(ru_emb[word])
            except:
                #if word[0].isupper:
                #    l.append(ru_emb[word[0] + morph.parse(word)[0].normal_form[1:]])
                #else:
                #    l.append(ru_emb[morph.parse(word)[0].normal_form])
                pass
    return l
```

```python
emb_df = pd.DataFrame(data_df['Question'].apply(lambda w: pd.Series(np.mean(word_prepare(w), axis=0))))
```

```python
emb_df['mean'] = pd.DataFrame(data_df['Question'].apply(lambda w: pd.Series(np.mean(np.mean(word_prepare(w), axis=0)))))
```

```python
punctuation = ['.',',',':',';','!','?','(',')','/','\\','|','-', '_']
quotes = ['\n', '\'', '\"']
special_quotes = ['«', '»']
numeral = '1234567890'
eng = 'abcdefghijklmnopqrstuvwxyz'
calculate = ['+', '=', '*']


def cnt_punctuation(string):
    return sum(list(map(lambda s: 1 if s in punctuation else 0, string)))

def cnt_special_quotes(string):
    return sum(list(map(lambda s: 1 if s in quotes else 0, string)))

def cnt_quotes(string):
    return sum(list(map(lambda s: 1 if s in special_quotes else 0, string)))

def cnt_numeral(string):
    return sum(list(map(lambda s: 1 if s in numeral else 0, string)))

def cnt_eng(string):
    return sum(list(map(lambda s: 1 if s in eng or s in eng.upper() else 0, string)))

def cnt_calculate(string):
    return sum(list(map(lambda s: 1 if s in calculate else 0, string)))

def cnt_whitespace(string):
    return sum(list(map(lambda s: 1 if s == ' ' else 0, string)))

#def cnt_dot(string):
#    return sum(list(map(lambda s: 1 if s == '.' else 0, string)))

def cnt_comma(string):
    return sum(list(map(lambda s: 1 if s == ',' else 0, string)))

def cnt_question(string):
    return sum(list(map(lambda s: 1 if s == '?' else 0, string)))

def cnt_character(string):
    return sum(list(map(lambda s: 1 if s != ' ' and s not in punctuation  and s not in special_quotes
                         and s not in numeral and s not in eng else 0, string)))

def cnt_upper(string):
    return sum(list(map(lambda s: 1 if s.isupper() else 0, string)))

def cnt_len(string):
    return len(string)

data_df['cnt_punctuation'] = data_df['Question'].apply(cnt_punctuation)
data_df['cnt_numeral'] = data_df['Question'].apply(cnt_numeral)
data_df['cnt_eng'] = data_df['Question'].apply(cnt_eng)
data_df['cnt_calculate'] = data_df['Question'].apply(cnt_calculate)
data_df['cnt_whitespace'] = data_df['Question'].apply(cnt_whitespace)
#data_df['cnt_dot'] = data_df['Question'].apply(cnt_dot)
#data_df['cnt_comma'] = data_df['Question'].apply(cnt_comma)
data_df['cnt_question'] = data_df['Question'].apply(cnt_question) 
data_df['cnt_character'] = data_df['Question'].apply(cnt_character)
data_df['cnt_upper'] = data_df['Question'].apply(cnt_upper)
data_df['cnt_len'] = data_df['Question'].apply(cnt_len)

data_df['cnt_punctuation_div_character'] = data_df['cnt_punctuation'] / data_df['cnt_character']
data_df['cnt_numeral_div_character'] = data_df['cnt_numeral'] / data_df['cnt_character']
data_df['cnt_eng_div_character'] = data_df['cnt_eng'] / data_df['cnt_character']
data_df['cnt_upper_div_character'] = data_df['cnt_upper'] / data_df['cnt_character']
data_df['cnt_whitespace_div_character'] = data_df['cnt_whitespace'] / data_df['cnt_character']
data_df['cnt_len_div_character'] = data_df['cnt_len'] / data_df['cnt_character']
#data_df['cnt_dot_div_whitespace_mult_len'] = data_df['cnt_dot'] / data_df['cnt_whitespace'] * data_df['cnt_len']
#data_df['cnt_comma_div_whitespace_mult_len'] = data_df['cnt_comma'] / data_df['cnt_whitespace'] * data_df['cnt_len']
#data_df['cnt_hyphen_div_whitespace_mult_len'] = data_df['cnt_hyphen'] / data_df['cnt_whitespace'] * data_df['cnt_len']
```

```python
data_df['clever_flg'] = data_df['Question'].apply(lambda s: 1 if 'клевер' in str(s).lower() else 0)
data_df['what_flg'] = data_df['Question'].apply(lambda s: 1 if 'что ' in str(s).lower() or 'кто' in str(s).lower()
                                                or 'кем ' in str(s).lower() or 'как' in str(s).lower() 
                                                or 'скольк' in str(s).lower() or 'назов' in str(s).lower() else 0)
data_df['is_flg'] = data_df['Question'].apply(lambda s: 1 if 'является' in str(s).lower() or 'был' in str(s).lower() 
                                              or 'есть' in str(s).lower() or 'такое' in str(s).lower() else 0)
data_df['in_flg'] = data_df['Question'].apply(lambda s: 1 if 'в ' in str(s).lower() or 'из ' in str(s).lower() 
                                              or 'между' in str(s).lower() else 0)
data_df['th_flg'] = data_df['Question'].apply(lambda s: 1 if 'это' in str(s).lower() else 0)
data_df['dots_flg'] = data_df['Question'].apply(lambda s: 1 if '...' in str(s).lower() else 0)
data_df['th2_flg'] = data_df['Question'].apply(lambda s: 1 if 'как ещ' in str(s).lower() else 0)
data_df['no_flg'] = data_df['Question'].apply(lambda s: 1 if ' не ' in str(s).lower() else 0)
```

```python
tokenizer = nltk.tokenize.SpaceTokenizer()
def word_morphy(sentence):
    words_dict = tokenizer.tokenize(sentence)
    l = []
    for word in words_dict:
        try:
            l.append(morph.parse(word)[0].score)
        except:
            pass
    return np.mean(l)
data_df['morphy_score'] = data_df['Question'].apply(word_morphy)
```

```python
%%time
# Чем больше P, тем меньше мы хотим штрафовать за разницу между train  и test
p = 0.8
# k - количество итераций
k = 75

cat = []

skf = KFold(n_splits=3, shuffle=True, random_state=7)

def score(params):
    print('Training with params:')
    print(params)
    w=[]
    best_iter = []
    

    for train_index, val_index in skf.split(X_train, y_train):
        x_train_1, x_valid_1 = X_train.iloc[train_index, :], X_train.iloc[val_index, :]
        y_train_1, y_valid_1 = y_train.iloc[train_index], y_train.iloc[val_index]
        train_data = lgb.Dataset(x_train_1, label=y_train_1, categorical_feature=cat)
        val_data = lgb.Dataset(x_valid_1, label=y_valid_1, categorical_feature=cat, reference=train_data)
        gbm = lgb.train(params,
                        train_data,
                        valid_sets = [train_data, val_data],
                        valid_names=['train', 'val'],
                        num_boost_round = 15800,
                        verbose_eval = False, 
                        categorical_feature=cat
                       )
        w.append([gbm.best_score['train']['auc'], gbm.best_score['val']['auc']])
        best_iter.append(gbm.best_iteration)
    nrounds = np.mean(best_iter)
    print('best iter:', int(nrounds), 'all iter:', best_iter)
    res = list(np.mean(w, axis=0))
    print("\t auc train {0}, auc test {1}, dif {2}, \n\t final score {3} \n\n".format(res[0], res[1], np.power(np.square(res[0]-res[1]), p), -res[1]+np.power(np.square(res[0]-res[1]), p)))
    return {'loss': -res[1]+np.power(np.square(res[0]-res[1]), p), 'status': STATUS_OK, 
            'mean_auc_train': res[0], 'mean_auc_test': res[1], 'best_iter': int(nrounds)}

def optimize(trials):
    space = {
        #'max_depth': hp.choice('max_depth', [-1, 6, 7]),
        'max_depth': -1,
        'max_bin': scope.int(hp.uniform('max_bin', 20, 2500)),
        'num_leaves': scope.int(hp.uniform('num_leaves', 10, 34)),
        'min_data_in_leaf': scope.int(hp.uniform('min_data_in_leaf', 125, 3500)),
        'lambda_l1': hp.quniform('lambda_l1', 0, 8, 0.25),
        'lambda_l2': hp.quniform('lambda_l2', 0, 13, 0.25),
        'learning_rate': hp.quniform('learning_rate', 0.02 , 0.1, 0.005),
        'min_child_weight': hp.quniform('min_child_weight', 0.01, 0.95, 0.05),
        'feature_fraction': hp.quniform('feature_fraction', 0.5, 0.9, 0.1),
        'bagging_fraction': hp.quniform('bagging_fraction', 0.5, 0.9, 0.1),
        'metric': ('auc',),
        'boosting_type': 'gbdt',
        'objective': 'binary',
        'nthread': 8,
        'early_stopping_rounds': 10,
        'silent':1,
    }
    best = fmin(score, space, algo=tpe.suggest, trials=trials, max_evals=k)
    print(best)

trials = Trials()
optimize(trials)
```
